### Regional Analysis

```python
genjp=df.groupby('Genre')['JP_Sales'].sum()
genna=df.groupby('Genre')['NA_Sales'].sum()
geneu=df.groupby('Genre')['EU_Sales'].sum()
genot=df.groupby('Genre')['Other_Sales'].sum()
```

```python
genjp
```

```python
plt.figure(figsize=(10,6))
plt.plot(genjp, label='JP_Sales',linestyle='-', marker='o', color='blue')
plt.plot(genna, label='NA_Sales',linestyle='--', marker='h', color='green')
plt.plot(geneu, label='EU_Sales',linestyle='-.', marker='D', color='magenta')
plt.plot(genot, label='Other_Sales',linestyle=':', marker='D', color='yellow')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.title("Different genre Sales Across the world")
plt.xlabel("Genre")
plt.ylabel('Sales')
plt.show()
```

### From Above Grapgh it is clear that most of the sale is contributed from North America, it holds a record of top action Genre games sele

### Different regional sales on different platform

```python
plfmj=plfm['JP_Sales'].sum()
plfme=plfm['EU_Sales'].sum()
plfmn=plfm['NA_Sales'].sum()
plfmo=plfm['Other_Sales'].sum()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(plfmj, label='JP_Sales',linestyle='-', marker='o', color='blue')
plt.plot(plfmn, label='NA_Sales',linestyle='--', marker='h', color='green')
plt.plot(plfme, label='EU_Sales',linestyle='-.', marker='D', color='magenta')
plt.plot(plfmo, label='Other_Sales',linestyle=':', marker='D', color='yellow')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.title("Different Platform sales Across the World")
plt.xlabel("Genre")
plt.ylabel('Sales')
plt.show()
```

#### analysisng regional sales of platform we got to finalise that North America holds the record as we have discussed above that NA has sold most of the games and now it is same in case of platform too
#### We can conclude that North America has contributed most sales in gaming all around the world
